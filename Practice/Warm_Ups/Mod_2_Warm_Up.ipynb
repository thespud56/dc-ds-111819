Provide the code that will scrape the rental price of the last apartment/house listed on this craigslist website https://washingtondc.craigslist.org/search/apa

```python
import requests
from bs4 import BeautifulSoup
```

```python
scrape = requests.get('https://washingtondc.craigslist.org/search/apa')
print(scrape.status_code)
```

```python
soup = BeautifulSoup(scrape.content, 'html.parser')
```

```python
soup.prettify
```

```python
listings = soup.find('ul', class_="rows")
```

```python
apts = listings.findAll('li')
last_apt = apts[-1]
last_apt
```

```python
last_price = last_apt.find('span', {"class" : 'result-price'}).text
last_price
```

Write a function to find all possible unique permutations of the word 'Mississippi'.
word = 'Mississippi'
Bonus: Write the function in such a way that it can find all possible unique permutations of any word.


```python
import itertools
```

```python
word = 'Mississippi'
def parser(word):
    for w in word:
        permutations = set(itertools.permutations(word))
    return permutations
```

```python
parser(word)
```
